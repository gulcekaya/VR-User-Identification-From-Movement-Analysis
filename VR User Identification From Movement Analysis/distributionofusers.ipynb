```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file with the correct delimiter
file_name = 'initialsurvey.csv'  # Replace with your actual file name if necessary
data = pd.read_csv(file_name, delimiter=';')

# Set up the figure and axes for the subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Age distribution
age_counts, age_bins, patches = axs[0, 0].hist(data['Age'], bins=10, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Age Distribution')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Frequency')
for i in range(len(age_counts)):
    axs[0, 0].text(age_bins[i] + (age_bins[i+1] - age_bins[i]) / 2, age_counts[i], str(int(age_counts[i])), ha='center', va='bottom')

# Gender distribution
gender_counts = data['Gender'].value_counts()
axs[0, 1].bar(gender_counts.index, gender_counts.values, color='lightgreen', edgecolor='black')
axs[0, 1].set_title('Gender Distribution')
axs[0, 1].set_xlabel('Gender')
axs[0, 1].set_ylabel('Count')
for i, count in enumerate(gender_counts.values):
    axs[0, 1].text(i, count, str(count), ha='center', va='bottom')

# Previous VR experience
vr_experience_counts = data['Previous VR experience (1-5)'].value_counts().sort_index()
axs[1, 0].bar(vr_experience_counts.index, vr_experience_counts.values, color='lightcoral', edgecolor='black')
axs[1, 0].set_title('Previous VR Experience')
axs[1, 0].set_xlabel('Experience Level')
axs[1, 0].set_ylabel('Frequency')
for i, count in enumerate(vr_experience_counts.values):
    axs[1, 0].text(i+1, count, str(count), ha='center', va='bottom')

# Previous general gaming experience
gaming_experience_counts = data['Previous general gaming experience (1-5)'].value_counts().sort_index()
axs[1, 1].bar(gaming_experience_counts.index, gaming_experience_counts.values, color='lightblue', edgecolor='black')
axs[1, 1].set_title('Previous General Gaming Experience')
axs[1, 1].set_xlabel('Experience Level')
axs[1, 1].set_ylabel('Frequency')
for i, count in enumerate(gaming_experience_counts.values):
    axs[1, 1].text(i+1, count, str(count), ha='center', va='bottom')

# Adjust layout
plt.tight_layout()
plt.show()
```
